```python
# Step (1): Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Step (2): Loading iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Step (3): Dropping the "species" column
iris = iris.drop(labels = 'species', axis = 1)
iris.head()
```

```python
# Step (4): Adding columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
# Step (5): Create dataframe with summary stats
summary_stats = pd.DataFrame({
    'mean' : iris.mean(),
    'std' : iris.std(),
    'min' : iris.min(),
    'max' : iris.max()
})

summary_stats = summary_stats.reindex(['sepal_length','sepal_width',
                                       'sepal_sum','petal_length',
                                       'petal_width','petal_sum'])

print(summary_stats)
```
